```python
import seaborn as sns
import matplotlib.pyplot as plt

#top 10 drugs with highest average AUC
top_drugs = df.groupby('DRUG_NAME')['AUC'].mean().sort_values(ascending=False).head(10).index

#filter the dataframe to include only those top drugs
df_top = df[df['DRUG_NAME'].isin(top_drugs)]

#create a side-by-side plot (boxplot and histogram)
plt.figure(figsize=(16, 6))

#boxplot
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='DRUG_NAME', y='AUC', data=df_top)
plt.yscale('logit')  # 🔥 Make y-axis log scale
plt.xticks(rotation=45)
plt.title("AUC Distribution of Top 10 Drugs (Log Y-axis)")
plt.xlabel("Drug Name")
plt.ylabel("Log AUC")

#histogram
plt.subplot(1, 2, 2)
sns.histplot(data=df_top, x='AUC', hue='DRUG_NAME', kde=True, element='step', multiple='stack')
plt.title("AUC Distribution of Top 10 Drugs (Histogram)")
plt.xlabel("AUC")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
